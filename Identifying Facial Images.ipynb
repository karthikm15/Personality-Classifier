```python
import collections
import os
from collections import OrderedDict
FACIAL_LANDMARKS_IDXS = OrderedDict([
    ("mouth", (48, 68)),
    ('inner_mouth', (60, 68)),
    ("right_eyebrow", (17, 22)),
    ("left_eyebrow", (22, 27)),
    ("right_eye", (36, 42)),
    ("left_eye", (42, 48)),
    ("nose", (27, 35)),
    ("jaw", (0, 17))
])
print(os.listdir('./data'))
```

```python
# import the necessary packages
from imutils import face_utils
import numpy as np
import argparse
import imutils
import dlib
import cv2
# construct the argument parser and parse the argument
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('./shape_predictor_68_face_landmarks.dat')

# load the input image, resize it, and convert it to grayscale
```

```python
for file in os.listdir('./data'):
    for filename in os.listdir('./data/' + file):
        filepath = './data/' + file + '/' + filename
        
        if (filepath[-5:] != '.xlsx'):
            for filename2 in os.listdir('./data/' + file + '/' + filename):
                image = cv2.imread('./data/' + file + '/' + filename + '/' + filename2)
                image = imutils.resize(image, width=750, height=750)
                gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                # detect faces in the grayscale image
                rects = detector(gray, 1)
#                 print(rects)

                print('./data/' + file + '/' + filename + '/' + filename2)
                stacked_coors = []
                for (index, rect) in enumerate(rects):
#                     print(index)
                    shape = predictor(gray, rect)
                    shape = face_utils.shape_to_np(shape)

                    a_dict = face_utils.FACIAL_LANDMARKS_IDXS.items()
                    tuple_list = list(a_dict)
                #     print(tuple_list)
        #             stacked_coors.append({"mouth" : [], # mouth
        #                      "inner_mouth" : [],
        #                      "right_eyebrow" : [], # right_eyebrow
        #                      "left_eyebrow" : [], # left_eyebrow
        #                      "right_eye" : [], # right_eye
        #                      "left_eye" : [], # left_eye
        #                      "nose" : [], # nose
        #                      "jaw" : []})
                    for f in range(len(tuple_list)):
                        key_value = tuple_list[f]
#                         print(key_value)
                        name = key_value[0]

                        i = key_value[1][0]
                        j = key_value[1][1]

                        for (x, y) in shape[i:j]:
                            stacked_coors.append((x,y))

                if (len(stacked_coors) == 76):
                    print("IT HAPPENED!")
                    if (not os.path.exists('./refined_images/' + file)):
                        os.makedirs('./refined_images/' + file)

                    cv2.imwrite('./refined_images/' + file + '/' + filename2, image)
                else: 
                    if (not os.path.exists('./discarded_images/' + file)):
                        os.makedirs('./discarded_images/' + file)

                    cv2.imwrite('./discarded_images/' + file + '/' + filename2, image)
        
        
```
