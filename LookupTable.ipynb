```
import pandas as pd
path = '/content/gdrive/MyDrive/personality_datasets/'
df = pd.read_csv(path+'lookup.csv')
df['Attributes'] = df['Unnamed: 0']
del df['Unnamed: 0']
```

```
df
```

```
import cv2
import tensorflow as tf


feature_list = []

unique_traits = ['Long Nose', 'No Long Nose']
def prepare(filepath):
    IMG_SIZE = 50  # 50 in txt-based
    img_array = cv2.imread(filepath)  # read in the image, convert to grayscale
    new_array = cv2.resize(img_array, (250, 252))  # resize image to match model's expected sizing
    return new_array.reshape(-1, 250, 252, 1)  # return the image with shaping that TF wants.
new_model = tf.keras.models.load_model('/content/gdrive/MyDrive/personality_datasets/big_nose_model3/')

prediction = new_model.predict([prepare("/content/gdrive/MyDrive/personality_datasets/test_image_nose/" + 'images.jpeg')])
prediction_list = prediction.tolist()
prediction_list = prediction_list[0]
print(prediction_list)

max_value = max(prediction_list)
maximum = -1
for index in prediction_list:
    if (maximum < index):
        maximum = index
unique_index = unique_traits[prediction_list.index(maximum)]
feature_list.append(unique_index)
```

```
# After models provide predictions: assume feature_list = ['FEATURE', 'FEATURE', 'FEATURE']

# feature_list = ['Square Forehead', 'Long Nose']
feature_df = pd.DataFrame()

for feature in feature_list:
  entry = df.loc[df['Attributes'] == feature]
  feature_df = pd.concat([feature_df,entry])


values = feature_df.sum()[:-1]
values = sorted(values.items(), key=lambda x: float(x[1]))
highest = values[:3] + values[-3:]
highest
```
