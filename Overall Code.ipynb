```
# Importing the necessary libraries to run the model itself (mainly using Tensorflow and Keras models for running this convolutional neural network)
import pandas as pd
import os
import shutil 
import os.path
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import numpy as np

import numpy as np
from sklearn.utils.multiclass import unique_labels
import os

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import itertools
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras import Sequential
from keras.applications import VGG19 #For Transfer Learning
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import SGD,Adam
from keras.callbacks import ReduceLROnPlateau
from keras.layers import Flatten,Dense,BatchNormalization,Activation,Dropout
from keras.utils import to_categorical
from pathlib import Path
```

```
# Creating image data generators to load the data that is provided by these two separate folders.
# Some important factors to consider:
#  - Image Size: (250, 252)
#  - Shuffling both the train_generator and the test_generator (note that the test_generator simply contains the valid dataset)
#  - Class mode is chosen as categorical since there are only two classes that we are focusing on
#  - The images were converted to grayscale in order to remove an extraneous factor towards testing our model.

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
train_generator = train_datagen.flow_from_directory(
        '/content/gdrive/MyDrive/personality_datasets/train_full_nose2/',
        target_size=(250, 252),
        color_mode = "grayscale",
        batch_size=64,
        class_mode='categorical',
        shuffle=True)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = train_datagen.flow_from_directory(
        '/content/gdrive/MyDrive/personality_datasets/valid_full_nose2/',
        target_size=(250, 252),
        batch_size=64,
        color_mode="grayscale",
        class_mode='categorical',
        shuffle=True)
```

```
# The model itself - some important points to take notice of:
#  - After research, we found that the 'relu' and 'softmax' functions would fit our specifications the best for the activation functions.
#     - We notcied that in many cases, our model was classifying the same class for all of the testing images, and this seemed like a probable fix to this issue.
#  - Two convolutional layers, two pooling layers, and four dense layers were used in the making of this model. 
#     - We experimented with more simplistic models and found that it generalized the data too much.
#     - We haven't done that much exploration into more complex models (meaning more layers for the model), which could lead to higher accuracy.

cnn = tf.keras.models.Sequential()
cnn.add(tf.keras.layers.Conv2D(filters=48, kernel_size=3, activation='relu', input_shape=(250, 252, 1), padding='same'))
#cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
cnn.add(tf.keras.layers.MaxPool2D((2,2), strides=(2,2), padding='same'))
cnn.add(tf.keras.layers.Conv2D(filters=48, kernel_size=3, activation='relu', padding = 'same'))
# cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding = 'same'))
# cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', padding = 'same'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding='same'))
cnn.add(tf.keras.layers.Flatten())
cnn.add(tf.keras.layers.Dense(128, activation='relu'))
cnn.add(tf.keras.layers.Dense(64, activation='relu'))
cnn.add(tf.keras.layers.Dense(32, activation='relu'))
# cnn.add(tf.keras.layers.Dense(32, activation='relu'))
cnn.add(tf.keras.layers.Dense(2, activation='softmax'))
```

```
# Running the model itself. Some factors to take account of:
#  - A non-predefined optimizer was initialized in order to calibrate the learning rate to our specifications (we experimented with different learning rates on the magnitude of tens
#    in order to find the most optimied solution to our problem. However, we still feel that learning rate may play an issue into our model.
#  - Binary_crossentropy was chosen as the loss model as it seemed optimal since we are solving a binary classification problem.
#  - We settled for 15 epochs as that seemed not to underfit the model. We saw that the model was producing the same value for all the testing images, so we increased the number of epochs
#    from the standard ten to fifteen.
from keras.optimizers import Adam

opt = Adam(lr = 0.0001)
cnn.compile(optimizer=opt, loss="binary_crossentropy", metrics=["accuracy"])
cnn.fit(x=train_generator, validation_data=test_generator, epochs=15 )
cnn.save('/content/gdrive/MyDrive/personality_datasets/big_nose_model4')
```

```
# Testing the model:
#  - Note that the folder 'test_doublechin' was used to act as the training images in this scenario - since some of the CelebA dataset images had NaN values, we simply skipped over them.
#     - These images were already done such that they only contained the area around the nose using the dlib library.
#  - A prepare() function was used in order to quickly preprocess the data and convert it into a numpy array using the image specifications needed by the model
#  - We ran over 8000 images of the testing model and checked how many were classified as big_nose and no_big_nose
#     - This process was also completed in order to check the accuracy of the model in overall
#     - Relatively little experimentation has been done in real life in order to test this model; however, one issue that we potentially see is the bias of the dataset itself.

import matplotlib.pyplot as plt
import cv2
import os
import tensorflow as tf

path, dirs, files = next(os.walk('/content/gdrive/MyDrive/personality_datasets/test_doublechin'))
train_count = len(files)
print(train_count)


def prepare(filepath):
    IMG_SIZE = 50  # 50 in txt-based
    img_array = cv2.imread(filepath)  # read in the image, convert to grayscale
    new_array = cv2.resize(img_array, (250, 252))  # resize image to match model's expected sizing
    return new_array.reshape(-1, 250, 252, 1)  # return the image with shaping that TF wants.

unique_traits = ['big_nose', 'no_big_nose']
new_model = tf.keras.models.load_model('/content/gdrive/MyDrive/personality_datasets/my_model')

unique_pictures = {'big_nose' : [],
 'no_big_nose' : [],
}

real_pictures = {'big_nose' : [],
 'no_big_nose' : [],
}

correct = 0
total = 0
unique_list = []
for test_file in os.listdir("/content/gdrive/MyDrive/personality_datasets/test_doublechin"):
    # print(test_file)
    prediction = new_model.predict([prepare("/content/gdrive/MyDrive/personality_datasets/test_doublechin/" + test_file)])
    prediction_list = prediction.tolist()
    prediction_list = prediction_list[0]
    max_value = max(prediction_list)
    maximum = -1
    for index in prediction_list:
        if (maximum < index):
            maximum = index
    unique_index = unique_traits[prediction_list.index(maximum)]
    unique_list.append(unique_index)
    if (unique_index == 'double_chin'):
      unique_index = 1
    else:
      unique_index = -1
    
    list1 = data[data['Image_Name'] == test_file].Big_Nose.tolist()
    list1 = list1[0]

    if (unique_index==list1):
      correct += 1
    
    total += 1
    print(total)
    if (total > 8000):
      break
    # unique_pictures[unique_index].append("/content/gdrive/MyDrive/personality_datasets/test_doublechin/" + test_file)

print(correct/total)
print(unique_list)
from collections import Counter 
d = Counter(unique_list)
print(d['big_nose'])
```
